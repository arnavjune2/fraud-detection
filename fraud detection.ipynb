Dealing with unbalanced data 
Using Under-Sampling(Undersampling is a technique to balance uneven datasets by keeping all of the data in the minority class and decreasing the size of the majority class)

```python
legit_sample = legit.sample(n=492)
```

```python
balanced_data = pd.concat([legit_sample, fraud] , axis=0).reset_index()
```

```python
balanced_data.head()
```

```python
balanced_data.Class.value_counts()
```

```python
balanced_data.groupby('Class').mean()
```

spliting data into test and train set


```python
X = balanced_data.drop(columns='Class', axis=1)
y = balanced_data.Class
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,stratify = y)
```

model training 
i.e Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

model evaluation


```python
prediction = model.predict(X_test)
```

```python
accuracy_score(prediction , y_test)
```
